```python
### RNN
PATH ='data/intermediate_data/jasp_analysis/'
FILE = 'regret_t_test_delta_posterior.csv'
```

```python
# read csv file
data = pd.read_csv(os.path.join(PATH, FILE), index_col=0)
```

```python
sns.distplot(data, kde=True, bins=100, hist=True, color = 'grey')
plt.axvline(x=0, color='black', ymin=0., ymax=0.99, linestyle='--', linewidth = 2)
plt.ylabel('Posterior density')
plt.xlabel('Effect size ' + r'$\delta$')

plt.tight_layout()
plt.savefig('../p1_generalization/plots/figure_2_B.svg', dpi = 600)
```
