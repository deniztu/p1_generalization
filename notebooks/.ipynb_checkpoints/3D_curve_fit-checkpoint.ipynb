```python
import scipy.optimize as optimize
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

# allows rotation of plots
%matplotlib notebook
```

```python
# load data
df = pd.read_csv('3D_data.csv')
df.head()
```

```python
A = np.vstack([df.X, df.Y, df.Z])
x = np.transpose(A[:2,:])
y = A[2,:]

def func(data, a, b):
    return data[:,0]*data[:,1]*a + b

guess = (1,1)
params, pcov = optimize.curve_fit(func, x, y, guess)
print(params)
```

```python
pred = func(x, params[0], params[1])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df.X,df.Y,pred)
ax.scatter(df.X,df.Y,df.Z)
```
