```python
# CONFIG

PATH_RNN_DATA ='data/rnn_raw_data/'
RNN_FILE_NAME = 'lstm2_a2c_nh_48_lr_0_0001_n_u_p_0_5_ew_0_vw_0_5_dr_0_5_res_d_f_p_0_1_rt_con_a_4_n_300_te_50000_id_{}_test_b_daw_p_{}'

# RNN_FILE_NAME ="lstm2_rei_nh_48_lr_0_0001_n_u_p_0_5_ew_0_vw_0_5_dr_0_5_res_d_f_p_0_1_rt_con_a_4_n_300_te_50000_id_{}_test_b_daw_p_{}"

# number of RNN instances
N_IDS = 30
# number of random walk instances
N_WALKS = 3
# number of trials per episode
N_TRIALS = 300
# number of actions/bandits
N_ACTIONS = 4

# plotting settings
my_dot_size = 10
my_alpha = 1
```

```python
# function to get counter-based uncertainty of actions

def get_uncertainty(choices, n_trials = 300, n_actions = 4):

    unc = np.zeros([n_trials, n_actions])

    for t in range(n_trials):
        for a in range(n_actions):
            if np.array(choices!=a)[t]:
                unc[t,a] = unc[t-1,a]+1
            else:
                unc[t,a] =  0
    return(unc)
```

```python
def diagnostic_plots(df, id_, walk):

    # default values
    my_cols = ['tab:blue','tab:orange','tab:green','tab:purple']

    # get choices
    choices = df.choice.to_numpy()
    # extract rewards
    rewards = np.array(df.loc[:, [col.startswith('p_rew') for col in df.columns.values]])
    # get chosen rewards
    chosen_rewards = np.array([rewards[t,a] for t, a in enumerate(choices)])
    # get switch_trials
    is_switch =  df.is_switch.to_numpy()==1

    # plotting defaults
    plt.rcParams["figure.figsize"] = (20, 20)
    fig, (ax1, ax2, ax3) = plt.subplots(3)
    fig.suptitle('ID {} Walk {}'.format(id_, walk))

    # create color for actions that led to punishment (negative reward)
    is_punishment = chosen_rewards < np.mean(rewards)
    punish_cols = ['white', 'black']
    punish_col = []
    for _, col in enumerate(is_punishment):
        punish_col.append(punish_cols[int(col)])
    punish_col = np.array(punish_col)

    ##################################################
    # predicted value of action a given PCA(H) 
    ##################################################

    # extract hidden unit activity
    hidden_units = df.loc[:, [col.startswith('rnn_state') for col in df.columns.values]]
    # init pca object
    pca = PCA()
    # standardize hidden unit activity
    st_hidden = StandardScaler().fit_transform(hidden_units)  
    # predictor X is principal components
    X = pca.fit_transform(st_hidden)

    # container for predicted action values
    action_value_preds = np.zeros([N_TRIALS, N_ACTIONS])

    for a in range(4):

        # get reward value of action a
        y = rewards[:,a]
        # create multinomial model
        model = LinearRegression().fit(X,y)
        # create prediction
        action_value_preds[:,a] = model.predict(X) 

    # get predicted action values of chosen actions
    chosen_action_value_preds =  np.array([action_value_preds[t,a] for t, a in enumerate(choices)])

    ##############################
    # predicted value plot (ax1) #
    ##############################

    # show dots for actions in the following y-positions
    x=[0.925, 0.95, 0.975, 1]
    actions = np.array([x[i] for i in choices])
    actions[np.invert(is_switch)] = actions[np.invert(is_switch)]-1

    # create color for actions
    tmp_col = []
    for col in range(choices.shape[0]):
        tmp_col.append(my_cols[int(choices[col])])

    # plot lines
    for a in range(4):
        ax1.plot(action_value_preds[:, a], alpha = 0.3, color = my_cols[a])   

    # plot scatters
    ax1.scatter(range(len(actions)), actions, c = tmp_col, s =40)
    ax1.scatter(np.array(range(300))[is_switch], chosen_action_value_preds[is_switch], c = np.array(tmp_col)[is_switch], s =70
                , edgecolors=punish_col[is_switch])
    ax1.scatter(np.array(range(300))[np.invert(is_switch)], actions[np.invert(is_switch)], c = np.array(tmp_col)[np.invert(is_switch)], s =40,
               edgecolors = punish_col[np.invert(is_switch)])

    # mean line
    ax1.axhline(y=np.mean(rewards), color='r', linestyle='-')
    # ylabel
    ax1.set(ylabel= 'Action Values | Hidden Units')

    #####################
    # running reward
    #####################

    '''
    running reward is the last observed reward of a given action
    '''

    # create running reward array
    run_rew = np.zeros([300, 4])

    for a in range(N_ACTIONS):

        for t in range(N_TRIALS):
            if np.array(choices==a)[t]:
                # if action is chosen update running reward
                run_rew[t, a] = rewards[t, a]
            else:
                # if not chosen running reward of an action remains unchanged
                run_rew[t, a] = run_rew[t-1, a]

    # get shifted running reward (as running reward t-1 informs action t)
    shifted_run_rew = np.zeros([300, 4])
    for a in range(4):
        shifted_run_rew[:, a] = np.array(pd.Series(run_rew[:,a]).shift(1))

    # get chosen shifted running reward (running reward t-1 of action t)
    chosen_shifted_run_rew = np.array([shifted_run_rew[t,a] for t, a in enumerate(choices)])

    ##############################
    # runnin reward plot (ax2)   #
    ##############################

    #show dots for actions in the following y-positions
    x=[1.025, 1.05, 1.075, 1.1]
    actions = np.array([x[i] for i in choices])
    actions[np.invert(is_switch)] = actions[np.invert(is_switch)]-1.15


    # plot lines
    for a in range(4):
        ax2.plot(shifted_run_rew[:, a], alpha = 1, color = my_cols[a])


    #create vertical lines
    ax2.vlines(x = np.array(range(300))[is_switch]
               , ymin =  chosen_shifted_run_rew[is_switch]
               , ymax = actions[is_switch]
              , linewidth = 0.3
              , color = 'gray')


    # plot scatters
    ax2.scatter(range(len(actions)), actions, c = tmp_col, s =40)
    ax2.scatter(np.array(range(300))[np.invert(is_switch)],
                actions[np.invert(is_switch)],
                c = np.array(tmp_col)[np.invert(is_switch)],
                s =40,
               edgecolors = punish_col[np.invert(is_switch)])

    ax2.scatter(np.array(range(300))[is_switch],
                chosen_shifted_run_rew[is_switch],
                c = np.array(tmp_col)[is_switch],
                s = 70,
                edgecolors=punish_col[is_switch])

    # mean line
    ax2.axhline(y=np.mean(rewards), color='r', linestyle='-') 
    # ylabel
    ax2.set(ylabel= 'Running Reward')


    ###################################
    # runnin reward rank plot (ax3)   #
    ###################################

    # running reward ranks with average on binding ranks
    # this results that the average rr_rank is 1.5 if choice is random
    ranked_rr = np.array([rankdata(shifted_run_rew[row,:], method = 'average')-1 for row in range(np.shape(shifted_run_rew)[0])])
    ranked_rr[np.isnan(shifted_run_rew)] = np.NAN  
    
    # chosen ranked rr
    chosen_ranked_rr = np.array([ranked_rr[t,a] for t, a in enumerate(choices)])   

    # plot lines
    for a in range(4):
        ax3.plot(ranked_rr[:,a] , linewidth = 3, alpha = 0.5, color = my_cols[a])

    #show dots for actions in the following y-positions
    x=[4.025, 4.05, 4.075, 4.1]
    actions = np.array([x[i] for i in choices])
    actions[np.invert(is_switch)] = actions[np.invert(is_switch)]-5.30

    #create vertical lines
    ax3.vlines(x = np.array(range(300))[is_switch]
            , ymin =  chosen_ranked_rr[is_switch]
               , ymax = actions[is_switch]
              , linewidth = 0.3
              , color = 'gray')

    # plot scatters  
    ax3.scatter(range(len(actions)), actions, c = tmp_col, s =40)

    ax3.scatter(np.array(range(300))[np.invert(is_switch)],
                actions[np.invert(is_switch)],
                c = np.array(tmp_col)[np.invert(is_switch)],
                s =40,
               edgecolors = punish_col[np.invert(is_switch)])

    ax3.scatter(np.array(range(300))[is_switch], chosen_ranked_rr[is_switch]
                , c = np.array(tmp_col)[is_switch], s =200, edgecolors=np.array(punish_col)[is_switch])

    # ylabel
    ax3.set(ylabel= 'Rank Running Reward (3 is Best)')
    
    #############
    # mean rank running reward:
    #############  

    # drop running reward of previous choice and current choice
    prev_choice = pd.Series(choices).shift(1).values
    shifted_run_rew_at_switch = shifted_run_rew[is_switch] 
    shifted_run_rew_other_at_switch = np.zeros([np.sum(is_switch), 2])
    for t in range(1, len(prev_choice[is_switch])):
        shifted_run_rew_other_at_switch[t,:] = np.delete(shifted_run_rew_at_switch[t,:],
                                           [int(prev_choice[is_switch][t]),
                                            int(choices[is_switch][t])])
        
    # run rew of switch target on 0 index
    chosen_rr_st = np.reshape(chosen_shifted_run_rew[is_switch], [-1,1])
    # concetenate uncertainty of switch target at index 0 and other on index 1 and 2
    cleaned_rr = np.hstack([chosen_rr_st, shifted_run_rew_other_at_switch])
    # rank uncertainty with bindings to average
    ranked_rr = np.array([rankdata(cleaned_rr[row,:], method = 'average')-1 for row in range(np.shape(cleaned_rr)[0])])
    # get mean rew rank rank at switch
    mean_rr_rank = np.mean(ranked_rr[:,0])

    
    #############
    # delta: running reward of switch target - mean running reward of other potential switch targets
    #############
    
    '''
    delta should be positive if switch targets more likely to the bandit with the highest running reward
    '''

    # run rew of switch target on 0 index
    chosen_rr_st = chosen_shifted_run_rew[is_switch]
    # mean of other switch options
    mean_other = np.mean(shifted_run_rew_other_at_switch, axis = 1)
    # get mean difference of switch_target rr and mean other at switch
    delta_rr = np.mean(chosen_rr_st - mean_other)
    
    ############
    # mean relative uncertainty of switch targets
    ############
    
    # get uncertainty
    unc = get_uncertainty(choices)
    # shift uncertainty (as uncertainty before a switch is relevant)
    shifted_unc = np.zeros_like(unc)
    for a in range(4):
        shifted_unc[:, a] = np.array(pd.Series(unc[:,a]).shift(1))

    # get chosen shifted uncertainty
    chosen_shifted_unc = np.array([shifted_unc[t,a] for t, a in enumerate(choices)])

    # drop unc of stay choices
    for i, is_stay in enumerate(np.invert(is_switch)):
        if is_stay:
            shifted_unc[i, choices[i]] = np.NAN

    # drop unc of switch choices
    for i, switch in enumerate(is_switch):
        if switch:
            shifted_unc[i, choices[i]] = np.NAN 

    # get max uncertainty of other options
    max_other = np.nanmax(shifted_unc, axis = 1)

    # get difference of chosen and max other uncertainty
    delta_unc = np.mean(chosen_shifted_unc[is_switch] - max_other[is_switch])
    
    #####################
    ### mean uncertainty rank
    ######################
    
    # get uncertainty
    unc = get_uncertainty(choices)
    # shift uncertainty (as uncertainty before a switch is relevant)
    shifted_unc = np.zeros_like(unc)
    for a in range(4):
        shifted_unc[:, a] = np.array(pd.Series(unc[:,a]).shift(1))

    # drop uncertainty of previous choice and current choice
    prev_choice = pd.Series(choices).shift(1).values
    shifted_unc_at_switch = shifted_unc[is_switch] 
    shifted_unc_other_at_switch = np.zeros([np.sum(is_switch), 2])
    for t in range(1, len(prev_choice[is_switch])):
        shifted_unc_other_at_switch[t,:] = np.delete(shifted_unc_at_switch[t,:],
                                           [int(prev_choice[is_switch][t]),
                                            int(choices[is_switch][t])])

    # uncertainty of switch target on 0 index
    chosen_unc_st = np.reshape(chosen_shifted_unc[is_switch], [-1,1])
    # concetenate uncertainty of switch target at index 0 and other on index 1 and 2
    cleaned_unc = np.hstack([chosen_unc_st, shifted_unc_other_at_switch])
    # rank uncertainty with bindings to average
    ranked_unc = np.array([rankdata(cleaned_unc[row,:], method = 'average')-1 for row in range(np.shape(cleaned_unc)[0])])
    # get mean uncertainty rank at switch
    mean_unc_rank = np.mean(ranked_unc[:,0])
    
    return(delta_rr, mean_rr_rank, delta_unc, mean_unc_rank)
```

```python
rnn_file_path = PATH_RNN_DATA + RNN_FILE_NAME

delta_rr = []
rank_rr = []
delta_unc = []
rank_unc = []

for walk in range(1, N_WALKS+1):
     for id_ in range(N_IDS):

        # get data from RNN instance
        d = pickle.load(open(rnn_file_path.format(id_, walk), 'rb'))
        df = d.reset_index()
        
        # plot diagnostics
        rr, mrr,  unc, m_r_unc = diagnostic_plots(df, id_ = id_, walk = walk)
        
        # append delta rr
        delta_rr.extend([rr])
        # append mean rr rank
        rank_rr.extend([mrr])
        # append delta unc
        delta_unc.extend([unc])
        # append uncertainty rank
        rank_unc.extend([m_r_unc])
```

```python
def my_violin_plot(ydata, ylabel, plot_name, note1 = '', note2 = ''):
    # create plot df
    plot_df = pd.DataFrame({"accuracy": ydata
                           })

    # half violin plots
    f, ax = plt.subplots()
    dy="accuracy"; ort="v"; 

    ax=pt.half_violinplot(y = dy, data = plot_df, bw = .2, cut = 0.,
                          scale = "area", width = .6, inner = None, orient = ort)

    ax=sns.stripplot(y = dy, data = plot_df, edgecolor = "white",
    size = 7, jitter = 1, zorder = 0, orient = ort, alpha = 0.8)

    x=sns.boxplot(y = dy, data = plot_df, color = "black", width = .15, zorder = 10,\
    showcaps = True, boxprops = {'facecolor':'none', "zorder":10},\
    showfliers=True, whiskerprops = {'linewidth':2, "zorder":10},\
    saturation = 1, orient = ort)
    
    plt.hlines(1, xmin = -0.5, xmax = 0.5, color = 'lightgreen', linewidth = 3, linestyles = '--')
    plt.annotate(note1, xy = [0.2,1.02], size = 14, color = 'lightgreen')
    plt.annotate(note2, xy = [0.2,0.97], size = 14, color = 'lightgreen')
    plt.annotate('', xy=(0.17, 1.05),
                 xycoords='data',
                 xytext=(0.17, 1.01),
                 textcoords='data',
                 arrowprops=dict(
                                 color='lightgreen',
                                 lw=1)
               )

    plt.annotate('', xy=(0.17, 0.95),
                 xycoords='data',
                 xytext=(0.17, 0.99),
                 textcoords='data',
                 arrowprops=dict(
                                 color='lightgreen',
                                 lw=1)
               )

    plt.ylabel(ylabel)
    plt.xlabel('')

    plt.tight_layout()
    
    plt.savefig('../p1_generalization/plots/{}.png'.format(plot_name), dpi = 600)
    
```
