```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
# data
X = [1, 3]
Y = [2, 4]
Z = [3, 5]
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(X,Y,Z, c = 'red', alpha = 1)

# x start, y start, z start,x end, y end, z end 
ax.quiver(X[0],Y[0],Z[0],X[1]-X[0],Y[1]-Y[0],Z[1]-Z[0])
```
