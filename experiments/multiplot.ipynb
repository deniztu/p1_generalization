# Figure 2

```python
# reading images
img = cv2.imread(os.path.join('plots','figure_2_A.png'))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
image1 = img

img = cv2.imread(os.path.join('plots','figure_2_B.png'))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
image2 = img

img = cv2.imread(os.path.join('plots','figure_2_C.png'))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
image3 = img

# img = cv2.imread(os.path.join('plots','figure_2_D.png'))
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# image4 = img

images = [image1, image2, image3]
```

```python
fig = plt.figure(figsize=(20, 15))
#plt.subplots_adjust(wspace = 1, hspace = 0.1)
for i, label in enumerate(('A', 'B', 'C')):
    ax = fig.add_subplot(1,3,i+1)
    ax.imshow(images[i])
    ax.axis('off')
    ax.text(0, 1, label, transform=ax.transAxes,
      fontsize=40, fontweight='bold', va='top', ha='right')

fig.tight_layout()
plt.savefig('../p1_generalization/plots/Figure_2.png', dpi = 600)
plt.show()
```
