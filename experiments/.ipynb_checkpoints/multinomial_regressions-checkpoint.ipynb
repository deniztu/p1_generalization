# Behavioural Analysis

```python
def create_x(data_file, path_to_data = 'data/rnn_raw_data/', stepsizes = np.linspace(1, 10, num=10)):
    
    # function outputs list of length len(stepsizes), which contain arrays of shifted by stepsize
    def create_lagged_arrays(array, stepsizes):
        res = list()
        for s in stepsizes:
            x = np.array(pd.DataFrame(array).shift(int(s))[0]) # fill 999 missing values later
            # reshape for fitting
            x = x.reshape(-1,1)
            res.append(x)
        return(res)
    
    d = pickle.load(open(path_to_data + data_file, 'rb'))
    df = d.reset_index()
    
    # create features and outcome
    choices = np.array(df.choice)
    rewards = np.array(df.reward)

    # features
    past_choices = create_lagged_arrays(choices, stepsizes)
    past_rewards = create_lagged_arrays(rewards, stepsizes)

    # fill na values
    past_choices = np.nan_to_num(past_choices, nan = 4) # will be excluded later in one-hot encoding
    past_rewards = np.nan_to_num(past_rewards, nan = 0)

    # outcome
    y = choices
    
    '''
    x is shape[n_trials, 9*len(stepsize)]

    # lag(1) choice 1,2,3,4, reward, choice*reward 1,2,3,4 | lag(2) choice 1,2,3,4, reward, choice*reward 1,2,3,4 | lag(3) .... 

    '''

    # create multinomial model
    model = LogisticRegression(multi_class='multinomial', solver = 'lbfgs')

    # create one-hot encoder
    ohe = OneHotEncoder(sparse=False)

    # stack lagged arrays horizontally
    for s,_ in enumerate(stepsizes):

        # one-hot encoding of choices
        past_choices_oh = ohe.fit_transform(past_choices[s].reshape(-1, 1))
        past_choices_oh = past_choices_oh[:,:4] # ohe for choice 4 is excluded

        if s == 0:
            # create x
            x = np.hstack((past_choices_oh, past_rewards[s].reshape(-1,1)))
        else:
            # add to x
            x = np.hstack((x, past_choices_oh))
            # add past_reward
            x = np.hstack((x, past_rewards[s].reshape(-1,1)))

        # add interaction (past_choices * past_rewards)
        x = np.hstack((x, past_choices_oh * past_rewards[s].reshape(-1,1)))

    # fit
    myfit = model.fit(x, y)
    
    print(myfit.summary())
    
    # get accuracy
    accuracy = model.score(x, y)
    
    # sort coefs to lags
    past_choice_coefs = np.zeros([4, 4, len(stepsizes)])
    past_reward_coefs = np.zeros([4,len(stepsizes)])
    past_interaction_coefs = np.zeros([4, 4, len(stepsizes)])

    for s, _ in enumerate(stepsizes):
        past_choice_coefs[:, :, s] = model.coef_[:,(9*s):(9*s + 4)]
        past_reward_coefs[:,s] = model.coef_[:,(9*s + 4)]
        past_interaction_coefs[:, :, s] = model.coef_[:,(5+9*s):(9*(s+1))]

    # return mean of diagonal elements coeff matrices of interaction and past_choice terms
    # e.g. p(choice i| past_choice = i) is diagonal in mentioned coef matrices
    # note! return mean of past reward coefs (this is the effect of obtained reward, regardless of choice)

    past_choice_diag_means = []
    past_reward_means = []
    past_interaction_diag_means = []

    for s, _ in enumerate(stepsizes):
        past_choice_diag_means.append(np.mean(np.diag(past_choice_coefs[:,:,s])))
        past_interaction_diag_means.append(np.mean(np.diag(past_interaction_coefs[:,:,s])))
        past_reward_means.append(np.mean(past_reward_coefs[:,s]))

    return past_choice_diag_means, past_reward_means, past_interaction_diag_means, accuracy
```

```python
# collecting data from all subject/instances

file = 'lstm2_a2c_nh_48_lr_0_0001_n_u_p_0_5_ew_0_vw_0_5_dr_0_5_res_d_f_p_0_1_rt_con_a_4_n_300_te_50000_id_{}_test_b_daw_p_{}'

past_choice_diag_means = []
past_reward_means = []
past_interaction_diag_means = []
accuracies = []


for id_ in range(0,20):
    for walk in range(1,4):
        
        a, b, c, acc= create_x(data_file = file.format(id_, walk), stepsizes = np.linspace(1, 10, num=10))
        past_choice_diag_means.append(a)
        past_reward_means.append(b)
        past_interaction_diag_means.append(c)
        accuracies.append(acc)
```

```python
# plotting

def plot_y_by_lag(coeff_array, title, ylab = 'Regression Weights'):
    coeff_array = np.transpose(coeff_array)
    stepsize = np.shape(coeff_array)[0]
    plt.plot(coeff_array,'-o', c = 'grey')
    plt.plot(np.mean(coeff_array, axis = 1),'-o', c = 'black')
    plt.ylabel(ylab)
    plt.xlabel('Lag (Trials)')
    plt.xticks(ticks = range(stepsize), labels = [int(i) for i in np.linspace(1, stepsize, num=stepsize)])
    plt.axhline(y=0, color='black', linestyle='--')
    plt.title(title)
```

```python
plot_y_by_lag(past_choice_diag_means, 'Choice Sensitivity')
```

```python
plot_y_by_lag(past_interaction_diag_means, 'Reward Sensitivity')
```

```python
plt.scatter(x = [0]*len(accuracies), y = accuracies)
plt.scatter(x = [0], y = np.mean(accuracies))
plt.ylabel('Accuracy')
```

```python
plot_y_by_lag(past_reward_means, 'Reward')
```

```python
def create_x1(df, stepsizes = np.linspace(1, 10, num=10)):
    
    # function outputs list of length len(stepsizes), which contain arrays of shifted by stepsize
    def create_lagged_arrays(array, stepsizes):
        res = list()
        for s in stepsizes:
            x = np.array(pd.DataFrame(array).shift(int(s))[0]) # fill 999 missing values later
            # reshape for fitting
            x = x.reshape(-1,1)
            res.append(x)
        return(res)
    
#     d = pickle.load(open(path_to_data + data_file, 'rb'))
#     df = d.reset_index()
    
    # create features and outcome
    choices = np.array(df.choice)
    rewards = np.array(df.reward)

    # features
    past_choices = create_lagged_arrays(choices, stepsizes)
    past_rewards = create_lagged_arrays(rewards, stepsizes)

    # fill na values
    past_choices = np.nan_to_num(past_choices, nan = 4) # will be excluded later in one-hot encoding
    past_rewards = np.nan_to_num(past_rewards, nan = 0)

    # outcome
    y = choices
    
    '''
    x is shape[n_trials, 9*len(stepsize)]

    # lag(1) choice 1,2,3,4, reward, choice*reward 1,2,3,4 | lag(2) choice 1,2,3,4, reward, choice*reward 1,2,3,4 | lag(3) .... 

    '''

    # create multinomial model
    model = LogisticRegression(multi_class='multinomial', solver = 'lbfgs', max_iter=300)

    # create one-hot encoder
    ohe = OneHotEncoder(sparse=False)

    # stack lagged arrays horizontally
    for s,_ in enumerate(stepsizes):

        # one-hot encoding of choices
        past_choices_oh = ohe.fit_transform(past_choices[s].reshape(-1, 1))
        past_choices_oh = past_choices_oh[:,:4] # ohe for choice 4 is excluded

        if s == 0:
            # create x
            x = np.hstack((past_choices_oh, past_rewards[s].reshape(-1,1)))
        else:
            # add to x
            x = np.hstack((x, past_choices_oh))
            # add past_reward
            x = np.hstack((x, past_rewards[s].reshape(-1,1)))

        # add interaction (past_choices * past_rewards)
        x = np.hstack((x, past_choices_oh * past_rewards[s].reshape(-1,1)))

    # fit
    model.fit(x, y)
    
    # get accuracy
    accuracy = model.score(x, y)
    
    # sort coefs to lags
    past_choice_coefs = np.zeros([4, 4, len(stepsizes)])
    past_reward_coefs = np.zeros([4,len(stepsizes)])
    past_interaction_coefs = np.zeros([4, 4, len(stepsizes)])

    for s, _ in enumerate(stepsizes):
        past_choice_coefs[:, :, s] = model.coef_[:,(9*s):(9*s + 4)]
        past_reward_coefs[:,s] = model.coef_[:,(9*s + 4)]
        past_interaction_coefs[:, :, s] = model.coef_[:,(5+9*s):(9*(s+1))]

    # return mean of diagonal elements coeff matrices of interaction and past_choice terms
    # e.g. p(choice i| past_choice = i) is diagonal in mentioned coef matrices
    # note! return mean of past reward coefs (this is the effect of obtained reward, regardless of choice)

    past_choice_diag_means = []
    past_reward_means = []
    past_interaction_diag_means = []

    for s, _ in enumerate(stepsizes):
        past_choice_diag_means.append(np.mean(np.diag(past_choice_coefs[:,:,s])))
        past_interaction_diag_means.append(np.mean(np.diag(past_interaction_coefs[:,:,s])))
        past_reward_means.append(np.mean(past_reward_coefs[:,s]))

    return past_choice_diag_means, past_reward_means, past_interaction_diag_means, accuracy
```

```python
def create_y(data_file, path_to_data = 'data/rnn_raw_data/', stepsizes = np.linspace(1, 10, num=10)):

#     file = 'lstm2_a2c_nh_48_lr_0_0001_n_u_p_0_5_ew_0_vw_0_5_dr_0_5_res_d_f_p_0_1_rt_con_a_4_n_300_te_50000_id_{}_test_b_daw_p_{}'
#     stepsizes = np.linspace(1, 10, num=10)
#     path_to_data = 'data/rnn_raw_data/'
#     data_file = file.format(0,1)

    # function outputs list of length len(stepsizes), which contain arrays of shifted by stepsize
    def create_lagged_arrays(array, stepsizes):
        res = list()
        for s in stepsizes:
            x = np.array(pd.DataFrame(array).shift(int(s))[0]) # fill 999 missing values later
            # reshape for fitting
            x = x.reshape(-1,1)
            res.append(x)
        return(res)

    d = pickle.load(open(path_to_data + data_file, 'rb'))
    df = d.reset_index()

    # create features and outcome
    choices = np.array(df.choice)
    rewards = np.array(df.reward)

    # features
    past_choices = create_lagged_arrays(choices, stepsizes)
    past_rewards = create_lagged_arrays(rewards, stepsizes)

    # fill na values
    past_choices = np.nan_to_num(past_choices, nan = 4) # will be excluded later in one-hot encoding
    past_rewards = np.nan_to_num(past_rewards, nan = 0)

    # outcome
    y = choices

    '''
    x is shape[n_trials, 9*len(stepsize)]

    # lag(1) choice 1,2,3,4, reward, choice*reward 1,2,3,4 | lag(2) choice 1,2,3,4, reward, choice*reward 1,2,3,4 | lag(3) .... 

    '''

    # create multinomial model
    model = LogisticRegression(multi_class='multinomial', solver = 'lbfgs')

    # create one-hot encoder
    ohe = OneHotEncoder(sparse=False)

    # stack lagged arrays horizontally
    for s,_ in enumerate(stepsizes):

        # one-hot encoding of choices
        past_choices_oh = ohe.fit_transform(past_choices[s].reshape(-1, 1))
        past_choices_oh = past_choices_oh[:,:4] # ohe for choice 4 is excluded

        if s == 0:
            # create x
            x = np.hstack((past_choices_oh, past_rewards[s].reshape(-1,1)))
        else:
            # add to x
            x = np.hstack((x, past_choices_oh))
            # add past_reward
            x = np.hstack((x, past_rewards[s].reshape(-1,1)))

        # add interaction (past_choices * past_rewards)
        x = np.hstack((x, past_choices_oh * past_rewards[s].reshape(-1,1)))

    # fit
    model.fit(x, y)

    # sort coefs to lags
    past_choice_coefs = np.zeros([4, 4, len(stepsizes)])
    past_reward_coefs = np.zeros([4,len(stepsizes)])
    past_interaction_coefs = np.zeros([4, 4, len(stepsizes)])

    for s, _ in enumerate(stepsizes):
        past_choice_coefs[:, :, s] = model.coef_[:,(9*s):(9*s + 4)]
        past_reward_coefs[:,s] = model.coef_[:,(9*s + 4)]
        past_interaction_coefs[:, :, s] = model.coef_[:,(5+9*s):(9*(s+1))]

    # return frobenius norm of coeff matrices of interaction and past_choice terms
    # e.g. p(choice i| past_choice = i) is diagonal in mentioned coef matrices
    # note! return l2-norm of past reward coefs (this is the effect of obtained reward, regardless of choice)

    past_choice_f_norms= []
    past_reward_l2_norms = []
    past_interaction_f_norms = []

    past_choice_f_norms.append([np.linalg.norm(past_choice_coefs[:,:,i]) for i in range(np.shape(past_choice_coefs)[2])])
    past_interaction_f_norms.append([np.linalg.norm(past_interaction_coefs[:,:,i]) for i in range(np.shape(past_interaction_coefs)[2])])
    past_reward_l2_norms.append([np.linalg.norm(past_reward_coefs[:,i]) for i in range(np.shape(past_reward_coefs)[1])])

    return past_choice_f_norms, past_reward_l2_norms, past_interaction_f_norms
```
